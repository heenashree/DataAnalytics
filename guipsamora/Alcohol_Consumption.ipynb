```python
import pandas as pd
df = pd.read_csv('drinks.csv')
df.head(10)
```

```python
#Which continent drinks more beer on average?
df1 = df.groupby('continent').beer_servings.mean()
df1
```

```python
#For each continent print the statistics for wine consumption

df1 = df.groupby('continent').total_litres_of_pure_alcohol.describe()
df1
```

```python
#Print the mean alcohol consumption per continent for every column



df.groupby('continent').total_litres_of_pure_alcohol.mean()
```
